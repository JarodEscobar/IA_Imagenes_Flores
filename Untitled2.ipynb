```python
import numpy as np
import os
import PIL
import PIL.Image
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

import os
from PIL import Image
```

```python
import pathlib
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file(origin=dataset_url,
                                   fname='flower_photos',
                                   untar=True)
data_dir = pathlib.Path(data_dir)

```

```python
image_count = len(list(data_dir.glob('*/*.jpg')))
print(image_count)
```

```python
roses = list(data_dir.glob('roses/*'))
PIL.Image.open(str(roses[0]))
```

```python
batch_size = 32
img_height = 180
img_width = 180
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
val_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
for image_batch, labels_batch in train_ds:
  print(image_batch.shape)
  print(labels_batch.shape)
  break
```

```python
num_classes = 5

model = tf.keras.Sequential([
  tf.keras.layers.Rescaling(1./255),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(num_classes)
])
```

```python
model.compile(
  optimizer='adam',
  loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['accuracy'])
```

```python
historial = model.fit(train_ds, validation_data=val_ds, epochs=4)  # Un entrenamiento de 3 a 5 es mas que suficiente creo yo
```

```python
def classify_image(image_path):
    if os.path.exists(image_path):
        # Cargar la imagen
        img = Image.open(image_path)

        # Mostrar la imagen seleccionada
        plt.imshow(img)
        plt.axis('off')  # Ocultar ejes
        plt.show()

        # Preprocesar la imagen para el modelo
        img = img.resize((img_height, img_width))  # Cambiar el tamaño a las dimensiones que espera el modelo
        img_array = image.img_to_array(img)  # Convertir la imagen a un array numpy
        img_array = np.expand_dims(img_array, axis=0)  # Añadir una dimensión extra para representar el batch
        img_array = img_array / 255.0  # Normalizar los valores de píxel entre 0 y 1

        # Predecir la clase con el modelo
        predictions = model.predict(img_array)
        predicted_class = np.argmax(predictions, axis=1)[0]

        # Mostrar la predicción
        print(f"Predicción: {class_names[predicted_class]}")
    else:
        print(f"La ruta {image_path} no existe. Verifica la ruta de la imagen.")
```
